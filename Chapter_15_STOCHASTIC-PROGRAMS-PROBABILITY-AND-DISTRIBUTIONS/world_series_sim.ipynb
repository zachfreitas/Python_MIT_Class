# World Series Simulation

```python
import random
import pylab

def playSeries(numGames, teamProb):
    numWon = 0
    for game in range(numGames):
        if random.random() <= teamProb:
            numWon += 1
    return (numWon > numGames//2)

def fractionWon(teamProb, numSeries, seriesLen):
    won = 0
    for series in range(numSeries):
        if playSeries(seriesLen, teamProb):
            won += 1
    return won/float(numSeries)

def simSeries(numSeries):
    prob = 0.5
    fracsWon, probs = [],[]
    while prob <= 1.0:
        fracsWon.append(fractionWon(prob, numSeries, 7))
        probs.append(prob)
        prob += 0.01
    pylab.axhline(0.95) # Draw a line at 95%
    pylab.plot(probs, fracsWon, 'k', linewidth = 3)
    pylab.xlabel('Probability of Winning a Game')
    pylab.ylabel('Probability of Winning a Series')
    pylab.title(str(numSeries) + ' Seven-Game Series')
```

```python
simSeries(400)
```

Notice that for the better team to win 95% of the time (0.95 on the y-axis), \
it needs to be so much better that it would win more than three out of every four games between the two teams. \
For comparison, in 2015, the two teams in the World Series had \
regular season winning percentages of 58.6% (Kansas City Royals) and 55.5% (New York Mets).
